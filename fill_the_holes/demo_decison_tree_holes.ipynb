# Preprocess, train and predict

## Data

```
from pyspark.sql import DataFrame

training_beers: DataFrame = spark.createDataFrame(
    [("Quinarelle", 70, 1), ("Kriek", 10, 0)],
    ["name", "bitterness", "good"])
  
beers.show()
```

```
beers_for_predictions: DataFrame = spark.createDataFrame(
    [("Guiness", 70)],
    ["name", "bitterness"])
  
beers_for_predictions.show()
```

##Preprocessing

Link: https://spark.apache.org/docs/latest/ml-features.html#vectorassembler

```
from pyspark.ml.feature import VectorAssembler

ready_training_beers: DataFrame = VectorAssembler(inputCols=["x"], outputCol="y").transform(training_beers)
  
ready_training_beers.show()
```

```
from pyspark.ml.feature import VectorAssembler

ready_beers_for_predictions: DataFrame = VectorAssembler(inputCols=["x"], outputCol="y").transform(beers_for_predictions)
  
ready_beers_for_predictions.show()
```

## Training

Link: https://spark.apache.org/docs/latest/ml-classification-regression.html#decision-tree-classifier

```
from pyspark.ml.classification import DecisionTreeClassifier, DecisionTreeClassificationModel

decision_tree_builder: DecisionTreeClassifier = DecisionTreeClassifier(labelCol="x", featuresCol="y")

model: DecisionTreeClassificationModel = decision_tree_builder.function_to_replace(ready_training_beers)
```

##Prediction

Link: https://spark.apache.org/docs/latest/ml-classification-regression.html#decision-tree-classifier

```
predictions: DataFrame = model.transform(ready_beers_for_predictions)

predictions.show()
```
